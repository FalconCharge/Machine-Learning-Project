```python

# Load dataset
data = pd.read_csv("../Raw/Student-por.csv", sep=";")

# Keep only selected columns
selected_cols = [
    "traveltime",  # 13
    "studytime",   # 14
    "failures",    # 15
    "schoolsup",   # 16
    "paid",        # 18
    "activities",  # 19
    "higher",      # 21
    "internet",    # 22
    "freetime",    # 25
    "absences",    # 30
    "G1",          # 31
    "G2",          # 32
    "G3"           # 33 (regression target)
]

df = data[selected_cols].copy()

# Add classification target: pass if G3 >= 10, else fail
df["pass_fail"] = df["G3"].apply(lambda x: "pass" if x >= 10 else "fail")

print("=== Dataset Overview (Selected Columns) ===")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# Column types summary
print("\nColumn Types:")
print(df.dtypes.value_counts())

# Missingness
missing_summary = (
    df.isnull().mean()
    .sort_values(ascending=False)
    .head(5)
    .apply(lambda x: f"{x:.2%}")
)
print("\nTop 5 Columns with Missing Values:")
print(missing_summary)

# Classification target distribution
print("\nClassification Target Distribution (pass/fail):")
print(df["pass_fail"].value_counts(normalize=True).map("{:.2%}".format))

# Regression target stats
print("\nRegression Target (G3) Summary:")
print(df["G3"].describe())

# Preview of the cleaned dataset
print("\nSample rows:")
print(df.head(5).to_string(index=False))
```
